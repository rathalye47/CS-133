In this hands-on exercise, you answer the following questions by using pandas data structures and methods to analyze the eukaryote genome data store in the following tab-delimited file. https://raw.githubusercontent.com/csbfx/advpy122-data/master/euk.tsv

```python
import pandas as pd
%matplotlib inline

# this input file is tab-delimited instead of comma-delimited
tsvFile = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/euk.tsv"
# Load the csv file into a dataframe
euk = pd.read_csv(tsvFile, sep="\t")
euk
```

## Q1. How many Mammals have at least 20,000 genes? What are their scientific names? 
*Note:* 
- *Mammals are under Class*
- *Scientific names are under Species*


```python
# Converts values in Number of genes column to numeric and converts any non-numeric values to NaN.
euk["Number of genes"] =  pd.to_numeric(euk["Number of genes"], errors='coerce')

# Mammals that have at least 20,000 genes.
print(euk[(euk['Class'] == 'Mammals') & (euk['Number of genes'] >= 20000)])
print("\n")

# Species that are Mammals with at least 20,000 genes.
print(euk[(euk['Class'] == 'Mammals') & (euk['Number of genes'] >= 20000)]['Species'])
print("\n")

# Converts the Series above to a list.
speciesList = euk[(euk['Class'] == 'Mammals') & (euk['Number of genes'] >= 20000)]['Species'].to_list()

# Gets the top 10 Species in the list.
speciesList[0:10]
```

### Q2. Animals are a part of Kingdom. How many records are there for each Class of Animals?

```python
# Calculates how many records there are for each Class of Animals.
euk[euk['Kingdom'] == 'Animals'].Class.value_counts()
```

### Q3. Animals are a part of Kingdom. How many unique Species are there for each Class of Animals?

```python
# Selects eukaryote genome data for the Animals Kingdom.
animals = euk[euk['Kingdom'] == 'Animals']

# Iterates over a list of unique Class names under the Animals Kingdom and counts the number of unique Species names for each Class of Animals.
for animal_class in animals.Class.unique():
  print(animal_class, animals[animals.Class == animal_class].Species.nunique())
```

### Q4. What are the unique scientific names of Mammals with the genus name Macaca?
*Recall: the scientific name starts with genus followed by a space and then the species name. Example: Homo sapiens.*

```python
# Selects eukaryote genome data for the Mammals Class.
mammals = euk[euk.Class == 'Mammals']

# Extracts eukaroyte genome data for Mammals with the genus name Macaca.
macaca_data = mammals[mammals.Species.str.split(' ').str[0] == 'Macaca']

# Gets a list of unique scientific names of Mammals with the genus name Macaca.
list(macaca_data.Species.unique())
```

### Q5. Modify the Species names to only contain the scientific names  and create a new dataframe.

Some of the names in the Species column have more than two parts, such as `Emiliania huxleyi CCMP1516`. Create a new column `Species` that contains only the first two parts of name, such as `Emiliania huxleyi`. Combine this new Species column with `Kingdom`, `Class`, `Size (Mb)`, `Number of genes`, and `Number of proteins` and store this new dataframe as `df_species`.

Hint: Follow Q3 in Lecture 4 with a little twist. Instead of just getting the first element from the split results, you will get the first two elements using `.str[0:2]` which will give you a list. You can then use `.str.join(" ")` to change it back to a string.

```python
# Creates an empty dataframe 'df_species'.
df_species = pd.DataFrame()

# Creates a new column 'Species' that contains only the first two parts of the name and adds the new column to the dataframe.
species_list = euk.Species.str.split(' ').str[0:2].str.join(' ').to_list()
df_species["Species"] = species_list

# Combines the new Species column with other columns from the 'euk' dataframe.
df_species = df_species.join(euk[["Kingdom", "Class", "Size (Mb)", "Number of genes", "Number of proteins"]])
df_species
```

### Q6. Create a pie plot using pandas to show the number of unique Species in each Class of Animals using the new dataframe you created in Q5
Hint: First, create a new dataframe that contains the number of unique Species and the index is the corresponding Animals Class. Then, use that dataframe to plot the pie plot.
[Check out this documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.pie.html)

```python
# Selects eukaryote genome data for the Animals Kingdom.
animals = df_species[df_species['Kingdom'] == 'Animals']

# Lists to store unique Animals Classes names and number of unique Species in each Animals Class.
animal_classes = []
num_unique_species = []

# Iterates over a list of unique Animals Classes.
# Appends each unique Animals Class to a list and appends its corresponding number of unique Species to another list.
for animal_class in animals.Class.unique():
  animal_classes.append(animal_class)
  num_unique_species.append(animals[animals.Class == animal_class].Species.nunique())

# Creates a new dataframe that contains the number of unique Species and its corresponding Animals Class.
df_species_counts = pd.DataFrame({"Animals": animal_classes, "Number of Unique Species": num_unique_species})

# Sets index to Animals column.
df_species_counts.set_index('Animals', inplace=True)

df_species_counts
```

```python
# Creates a panda pie plot using the dataframe above.
df_species_counts.plot.pie(y='Number of Unique Species', figsize=(11,11))
```
