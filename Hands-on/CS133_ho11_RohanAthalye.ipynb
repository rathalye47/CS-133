Your instructor's favorite video game has been Mario Kart Wii, a racing game released for the Wii platform in 2008. And, IGN agrees with her that it is a great game -- their rating for this game is a whopping 8.9! Inspired by the success of this game, your instructor is considering creating your very own racing game for the Wii platform. Perform the following analyses to help her determine which platform she should focus on.

```python
ign_scores2 = ign_scores.reset_index() # Resets index and stores in a new dataframe.
plt.figure(figsize=(8,8)) # Sets the width and height of the plot.
plt.title("Xbox One Platform has the highest score for Racing games", y=1.01)
sns.barplot(data=ign_scores2, x="Racing", y="Platform"); # Creates the barplot.
plt.xlabel("Racing game scores");
```

#### Problem 4 - (2 points)
Since your instructor's gaming interests are pretty broad, you can help her decide to use the IGN scores to determine the choice of genre and platform. 

1. Create a heatmap using the IGN scores by genre and platform and include the scores in the cells of the heatmap.
2. Base on the heatmap, which combination of genre and platform receives the highest average ratings?  Which combination receives the lowest average rankings? Write the answers in a markdown cell.

```python
plt.figure(figsize=(10,8)) # Sets the width and height of the plot.
plt.title("IGN Scores for each Genre, by Platform") 
sns.heatmap(data=ign_scores, annot=True, cmap="GnBu") # Creates the heatmap.
plt.xlabel("Genre");
```

Based on the heatmap, the combination of the Simulation genre with the PlayStation 4 platform receives the highest average ratings. 2 combinations receive the lowest average ratings: the Fighting genre with the Game Boy Color platform and the Shooter genre with the Game Boy Color platform.

## Problem 5 - (4 points)
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
csv_file = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pokemon = pd.read_csv(csv_file, index_col="Name")
pokemon.head()
```

```python
# Filters the dataframe to only keep data with Type 1 equal to Water, Normal, Grass, Bug, and Psychic.
pokemon = pokemon[pokemon["Type 1"].isin(["Water", "Normal", "Grass", "Bug", "Psychic"])]

# Creates a series of colors corresponding to the Type 1 names.
type1_color = pokemon["Type 1"].replace({"Water": "blue", "Normal": "red", "Grass": "green", "Bug": "purple", "Psychic": "pink"})

# Extracts the first 25 rows of the dataframe.
pokemon = pokemon.head(25)

# Creates a clustermap with color.
g=sns.clustermap(
    pokemon.drop(columns=["#", "Type 1", "Type 2", "Generation", "Total", "Legendary"]),
    cmap="BuPu",
    col_cluster=False,
    figsize=(12,8),
    row_colors = type1_color
)
g.fig.suptitle(
    """HP, Attack, Defense, Special Attack, Special Defense, and Speed in Pokemon, colored by Type 1
       (blue=Water, red=Normal, green=Grass, purple=Bug, pink=Psychic)""", y=1.05);
```
